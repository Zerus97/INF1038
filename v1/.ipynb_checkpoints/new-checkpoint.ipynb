Removendo Outliers

```python
def remove_outliers(df, coluna):
    media = df[coluna].mean()
    desvio_padrao = df[coluna].std()
    
    limite_inferior = media - 3 * desvio_padrao
    limite_superior = media + 3 * desvio_padrao
    
    return df[(df[coluna] >= limite_inferior) & (df[coluna] <= limite_superior)]
```

```python
colunas_numericas = ['diametro_estimado_minimo', 'diametro_estimado_maximo', 'velocidade_relativa', 'distancia', 'magnitude_absoluta']

for coluna in colunas_numericas:
    df = remove_outliers(df, coluna)
```

As colunas diametro e velocidade tem correlação positiva frama. Mas tem negativa forte com magnitude absoluta. Quero entender como isso se relaciona com o perigoso.

```python
bins = 15

fig, axes = plt.subplots(3, 2, figsize=(10, 10))  # 3 linhas, 2 colunas

sns.set(style="whitegrid")

for i, coluna in enumerate(colunas_numericas):
    row = i // 2  
    col = i % 2 
    sns.histplot(df[coluna], bins=bins, kde=False, color='blue', edgecolor='black', ax=axes[row, col])
    axes[row, col].set_title(f'{coluna}', fontsize=10)
    axes[row, col].set_xlabel(coluna, fontsize=8)
    axes[row, col].set_ylabel('Frequência', fontsize=8)
    axes[row, col].tick_params(axis='x', labelsize=8)
    axes[row, col].tick_params(axis='y', labelsize=8)

fig.delaxes(axes[2, 1])

plt.tight_layout()
plt.show()
```

```python
# Calcular a participação de True e False
share = df['perigoso'].value_counts(normalize=True)

# Criar gráfico de pizza
plt.figure(figsize=(4, 3))
plt.pie(share, labels=share.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])
plt.title('Participação de True e False')
plt.axis('equal')  # Para garantir que o gráfico seja um círculo
plt.show()
```

Apenas 8,9% é perigoso

Multivariada

```python
matriz_correlacao = df[colunas_numericas].corr()

plt.figure(figsize=(6, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlação')
plt.show()
```

```python
variaveis_continuas = ['diametro_estimado_minimo', 'diametro_estimado_maximo', 'velocidade_relativa', 'distancia', 'magnitude_absoluta']

plt.figure(figsize=(15, 10))
for i, var in enumerate(variaveis_continuas, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='perigoso', y=var, data=df)
    plt.title(f'{var} x perigoso')
plt.tight_layout()
plt.show()
```
