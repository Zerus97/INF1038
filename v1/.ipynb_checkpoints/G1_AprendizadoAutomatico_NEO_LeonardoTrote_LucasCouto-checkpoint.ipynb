Removendo Outliers

```python
colunas_numericas = ['diametro_estimado_minimo', 'diametro_estimado_maximo', 'velocidade_relativa', 'distancia', 'magnitude_absoluta']
```

```python
def remove_outliers(df, colunas):
    df_sem_outliers = df.copy() 
    for coluna in colunas:
        media = df_sem_outliers[coluna].mean()
        desvio_padrao = df_sem_outliers[coluna].std()
        
        limite_inferior = media - 3 * desvio_padrao
        limite_superior = media + 3 * desvio_padrao
        
        df_sem_outliers = df_sem_outliers[(df_sem_outliers[coluna] >= limite_inferior) & 
                                           (df_sem_outliers[coluna] <= limite_superior)]
    
    return df_sem_outliers
```

```python
for coluna in colunas_numericas:
    df = remove_outliers(df, colunas_numericas)
```

```python
num_registros = df['id'].count() 
print(f"Número de registros: {num_registros}")
```

```python
perda = round((90836/77412 - 1)*100,2)

print(f"Perda após a remoção dos outliers: {perda} %")
```

```python
df_normalizado = df.copy()

for coluna in colunas_numericas:
    min_val = df[coluna].min()
    max_val = df[coluna].max()
    df_normalizado[coluna] = (df[coluna] - min_val) / (max_val - min_val)
```

```python
display(df_normalizado)
```

Univariada

As colunas diametro e velocidade tem correlação positiva frama. Mas tem negativa forte com magnitude absoluta. Quero entender como isso se relaciona com o perigoso.

```python
bins = 15

fig, axes = plt.subplots(3, 2, figsize=(10, 10))  # 3 linhas, 2 colunas

for i, coluna in enumerate(colunas_numericas):
    row = i // 2  
    col = i % 2 
    sns.histplot(df[coluna], bins=bins, kde=False, color='blue', edgecolor='black', ax=axes[row, col])
    axes[row, col].set_title(f'{coluna}', fontsize=15)
    axes[row, col].set_xlabel(coluna, fontsize=12)
    axes[row, col].set_ylabel('Frequência', fontsize=12)
    axes[row, col].tick_params(axis='x', labelsize=12)
    axes[row, col].tick_params(axis='y', labelsize=12)

fig.delaxes(axes[2, 1])

plt.tight_layout()
plt.show()
```

```python
share = df_normalizado['perigoso'].value_counts(normalize=True)

plt.figure(figsize=(4, 3))
plt.pie(share, labels=share.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])
plt.title('Participação de True e False')
plt.axis('equal')
plt.show()
```

Apenas 6,8% é perigoso

Multivariada

```python
matriz_correlacao = df[colunas_numericas].corr()

plt.figure(figsize=(8, 8))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar_kws={"shrink": .8})
plt.title('Matriz de Correlação')
plt.show()
```

```python
num_colunas = 2  
num_linhas = (len(colunas_numericas) + num_colunas - 1) // num_colunas

plt.figure(figsize=(15, 6 * num_linhas))

for i, coluna in enumerate(colunas_numericas):
    plt.subplot(num_linhas, num_colunas, i + 1)
    sns.boxplot(x='perigoso', y=coluna, data=df_normalizado)
    plt.title(f'Box Plot de {coluna} por Perigo')
    plt.xlabel('Perigoso (0 = Não, 1 = Sim)')
    plt.ylabel(coluna)
    plt.xticks([0, 1], ['Não Perigoso', 'Perigoso'])

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df_normalizado.drop(columns=['id']), hue='perigoso', palette={0: 'blue', 1: 'red'})
plt.title('Matriz de Dispersão - NEOs Perigosos')
plt.show()
```
