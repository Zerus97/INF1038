Não faz sentido. Em cidade tem nome de bairro, rua e outras pontos de ferência como parques e praças. Todos localizados em Londres. Foi um erro de preenchimento então podemos considerar que em todos as linhas a cidade é Londres.

```python
df['city'] = "London"
```

O dado original está em pés quadrados, mas para nós fica mais claro se estiver em metros quadrados. Então podemos converter.

```python
df['sqm'] = df['sqft'] * 0.092903
```

```python
unique_locations = df['location'].unique()

for location in unique_locations:
    print(location)
```

O campo location também tem o mesmo problema de preenchimento que cidade, mas nesse caso deveria ser um campo de mais granularidade. Existem campos iguais com espaços extras, nomes de bairros, ruas, etc. Precisamos reduzir o número de opções e padronizar. 

Como temos o campo Post Code que parece estar preenchido corretamente, podemos usá-lo em uma API que retorne informações sobre o local.

```python
import requests

#pega o bairro pelo post code
def get_neighborhood_from_postcode(postcode, index):
    print(f"Consultando o código postal na linha {index}: {postcode}")
    url = f"https://api.postcodes.io/postcodes/{postcode}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if data['status'] == 200:
            neighborhood = data['result'].get('admin_district', 'Bairro não encontrado')
            print(f"Linha {index}: Bairro encontrado - {neighborhood}")
            return neighborhood
        else:
            print(f"Linha {index}: Código postal não encontrado")
            return "Código postal não encontrado"
    else:
        print(f"Linha {index}: Erro ao recuperar dados")
        return "Erro ao recuperar dados"

# aplica a função na coluna post code
df['neighborhood'] = df.apply(lambda row: get_neighborhood_from_postcode(row['postal_code'], row.name), axis=1)
```

Como esse processo é demorado, vamos salvar como um arquivo novo e podemos trabalhar com ele.

```python
#salvando
df.to_csv('London_houses_cleaned.csv', index=False)
```
